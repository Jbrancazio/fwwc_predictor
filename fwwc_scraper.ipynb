## Define Scrubing function

```python
def get_stats_home(main_page):
    page= requests.get(main_page)
    soup = BeautifulSoup(page.content, 'html.parser')
    dates=soup.findAll("div", {"class":"fi-mu-list"})

    game_date=[]
    game_ids=[]
    team=[]
    venue=[]
    goals=[]
    attempts=[]
    on_target=[]
    off_target=[]
    blocked=[]
    corners=[]
    offsides=[]
    ball_possession=[]
    pass_accuracy=[]
    passes=[]
    passes_completed=[]
    distance=[]
    tackles=[]
    blocks=[]
    clearance=[]
    yellow_cards=[]
    direct_red_cards=[]
    indirect_red_cards=[]
    fouls=[]

    for date in dates:
        games=date.findAll("div", {"class":"fi-mu result fi-mu-national result"})

        for game in games:
            if game.has_attr("data-id"):
                game_ids.append(game["data-id"])

            date_times=game.findAll("div", {"class":"fi-mu__info__datetime"})

            for date_time in date_times:
                if date_time.has_attr("data-utcdate"):
                    game_date.append(np.datetime64(date_time["data-utcdate"]))


    # create the variable for the away team
    for game_id in game_ids:
        page = requests.get("https://www.fifa.com/womensworldcup/matches/match/"+str(game_id)+"/_libraries/_live_statistics?qs=1")
        soup = BeautifulSoup(page.content, 'html.parser')
        html = list(soup.children)[3]
        body = list(html.children)[3]
        game = list(html.children)[3]

        stats=game.findAll("tr",{"class":"table__stats-progress-bar__data"})

        statistics=[]

        for stat in stats:
            s=stat.findAll("td")
            for i in s:
                statistics.append(i.text)
                
        team.append(game.find("div", {"class":"team-home__inner-helper"}).text.replace("\n",""))
        venue.append("home")
        goals.append(float(statistics[0]))
        attempts.append(float(statistics[3]))
        on_target.append(float(statistics[6]))
        off_target.append(float(statistics[9]))
        blocked.append(float(statistics[12]))
        corners.append(float(statistics[15]))
        offsides.append(float(statistics[18]))
        ball_possession.append(float(statistics[21].replace("\n","").replace("%",""))/100)
        pass_accuracy.append(float(statistics[24].replace("\n","").replace("%",""))/100)
        passes.append(float(statistics[27]))
        passes_completed.append(float(statistics[30]))
        distance.append(float(statistics[33].replace(" km","")))
        tackles.append(float(statistics[36]))
        blocks.append(float(statistics[39]))
        clearance.append(float(statistics[42]))
        yellow_cards.append(float(statistics[45]))
        direct_red_cards.append(float(statistics[48]))
        indirect_red_cards.append(float(statistics[51]))
        fouls.append(float(statistics[54]))
        
    home_results=pd.DataFrame(zip(game_date,game_ids,team,venue,goals,attempts,on_target,off_target,blocked,corners,offsides,
                             ball_possession,pass_accuracy,passes,passes_completed,distance,tackles,blocks,
                             clearance,yellow_cards,direct_red_cards,indirect_red_cards,fouls),
                         columns=['game_date','game_ids', 'team', 'venue', 'goal', 'attempts', 'on_target','off_target',
                                  'blocked', 'corners', 'offsides', 'ball_possession','pass_accuracy', 'passes',
                                  'passes_completed', 'distance', 'tackles', 'blocks', 'clearance', 'yellow_cards',
                                  'direct_red_cards', 'indirect_red_cards', 'fouls'])
    return home_results
    
```

```python
def get_stats_away(main_page):
    page= requests.get(main_page)
    soup = BeautifulSoup(page.content, 'html.parser')
    dates=soup.findAll("div", {"class":"fi-mu-list"})

    game_date=[]
    game_ids=[]
    team=[]
    venue=[]
    goals=[]
    attempts=[]
    on_target=[]
    off_target=[]
    blocked=[]
    corners=[]
    offsides=[]
    ball_possession=[]
    pass_accuracy=[]
    passes=[]
    passes_completed=[]
    distance=[]
    tackles=[]
    blocks=[]
    clearance=[]
    yellow_cards=[]
    direct_red_cards=[]
    indirect_red_cards=[]
    fouls=[]

    for date in dates:
        games=date.findAll("div", {"class":"fi-mu result fi-mu-national result"})

        for game in games:
            if game.has_attr("data-id"):
                game_ids.append(game["data-id"])

            date_times=game.findAll("div", {"class":"fi-mu__info__datetime"})

            for date_time in date_times:
                if date_time.has_attr("data-utcdate"):
                    game_date.append(np.datetime64(date_time["data-utcdate"]))


    # create the variable for the away team
    for game_id in game_ids:
        page = requests.get("https://www.fifa.com/womensworldcup/matches/match/"+str(game_id)+"/_libraries/_live_statistics?qs=1")
        soup = BeautifulSoup(page.content, 'html.parser')
        html = list(soup.children)[3]
        body = list(html.children)[3]
        game = list(html.children)[3]

        stats=game.findAll("tr",{"class":"table__stats-progress-bar__data"})

        statistics=[]

        for stat in stats:
            s=stat.findAll("td")
            for i in s:
                statistics.append(i.text)
        team.append(game.find("div", {"class":"team-away__inner-helper"}).text.replace("\n",""))
        venue.append("away")
        goals.append(float(statistics[2]))
        attempts.append(float(statistics[5]))
        on_target.append(float(statistics[8]))
        off_target.append(float(statistics[11]))
        blocked.append(float(statistics[14]))
        corners.append(float(statistics[17]))
        offsides.append(float(statistics[20]))
        ball_possession.append(float(statistics[23].replace("\n","").replace("%",""))/100)
        pass_accuracy.append(float(statistics[26].replace("\n","").replace("%",""))/100)
        passes.append(float(statistics[29]))
        passes_completed.append(float(statistics[32]))
        distance.append(float(statistics[35].replace(" km","")))
        tackles.append(float(statistics[38]))
        blocks.append(float(statistics[41]))
        clearance.append(float(statistics[44]))
        yellow_cards.append(float(statistics[47]))
        direct_red_cards.append(float(statistics[50]))
        indirect_red_cards.append(float(statistics[53]))
        fouls.append(float(statistics[56]))
    away_results=pd.DataFrame(zip(game_date,game_ids,team,venue,goals,attempts,on_target,off_target,blocked,corners,offsides,
                             ball_possession,pass_accuracy,passes,passes_completed,distance,tackles,blocks,
                             clearance,yellow_cards,direct_red_cards,indirect_red_cards,fouls),
                         columns=['game_date','game_ids', 'team', 'venue', 'goal', 'attempts', 'on_target','off_target',
                                  'blocked', 'corners', 'offsides', 'ball_possession','pass_accuracy', 'passes',
                                  'passes_completed', 'distance', 'tackles', 'blocks', 'clearance', 'yellow_cards',
                                  'direct_red_cards', 'indirect_red_cards', 'fouls'])
    return away_results
```

```python
def get_results(tournament,main_page):
    home=get_stats_home(main_page)
    away=get_stats_away(main_page)
    
    results=home.append(away)
    results.sort_values(["game_date","venue"],inplace=True)
    results.reset_index(inplace=True)
    results.drop("index",axis=1,inplace=True)
    
    results.to_csv(tournament+".csv")
    return results
```
